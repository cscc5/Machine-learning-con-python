hallar las emisiones de CO2 en función del tamaño del motor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
df = pd.read_csv("./FuelConsumptionCo2.csv")
```

```python
regr.fit(df[["ENGINESIZE"]],df["CO2EMISSIONS"])
```

```python
print("m:", regr.coef_)
```

```python
print("b:", regr.intercept_)
```

```python
plt.scatter(df["ENGINESIZE"], df["CO2EMISSIONS"], color="blue")
plt.xlabel("Engine size")
plt.ylabel("Emissions CO2")
plt.show()
```

```python
print(np.corrcoef(df["ENGINESIZE"], df["CO2EMISSIONS"]))
```

```python
df.head()
```

```python
df["predict"]=regr.predict(df[["ENGINESIZE"]])
```

```python
df["residual"]=df["CO2EMISSIONS"]-df["predict"]
```

```python
df.head(20)
```

```python
RMSE = np.mean(df["residual"]**2)**(1/2)
print("Mi error RMSE: ", RMSE)
```
